```python
import pandas as pd

data = pd.DataFrame({
    "text": [
        # Phishing Emails
        "Click here to claim your free prize",
        "Your account has been suspended. Verify now",
        "Congratulations! You won a lottery",
        "Urgent: Verify your bank account immediately",
        "Claim your reward by clicking this link",
        "Update your password now to avoid suspension",
        "You have received a cashback offer",
        "Login to secure your account",
        "Your package delivery is pending. Click here",
        "Free gift card waiting for you",
        "Limited-time offer. Claim now",
        "Your account will be closed unless verified",
        "Win a brand new smartphone today",
        "Confirm your payment details immediately",
        "Security alert: login required now",
        "Exclusive reward for selected users",
        "Click to unlock your bonus reward",
        "Verify your identity to continue",
        "Your tax refund is ready. Claim it now",
        "Act now to prevent account suspension",

        # Safe Emails
        "Meeting scheduled for tomorrow",
        "Project report attached for review",
        "Please review the document and provide feedback",
        "Team meeting at 3 PM today",
        "Thank you for your application",
        "Monthly performance report attached",
        "Reminder for tomorrow presentation",
        "Invoice for your recent purchase attached",
        "Welcome to the organization",
        "Training session starts next week",
        "Please find the meeting agenda attached",
        "Your leave request has been approved",
        "Weekly project status update",
        "Client meeting rescheduled to Friday",
        "Happy birthday from the team",
        "Office will remain closed on Monday",
        "Kindly review the updated project plan",
        "Thank you for attending the workshop",
        "The document has been shared successfully",
        "Your registration has been confirmed"
    ],

    "label": [
        "Phishing","Phishing","Phishing","Phishing","Phishing",
        "Phishing","Phishing","Phishing","Phishing","Phishing",
        "Phishing","Phishing","Phishing","Phishing","Phishing",
        "Phishing","Phishing","Phishing","Phishing","Phishing",

        "Safe","Safe","Safe","Safe","Safe",
        "Safe","Safe","Safe","Safe","Safe",
        "Safe","Safe","Safe","Safe","Safe",
        "Safe","Safe","Safe","Safe","Safe"
    ]
})

print("Dataset Size:", len(data))
print(data.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix

X = data["text"]
y = data["label"]

vectorizer = TfidfVectorizer()
X_features = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_features,
    y,
    test_size=0.2,
    random_state=42
)

model = MultinomialNB()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

print("Accuracy:", round(accuracy_score(y_test, predictions) * 100, 2), "%")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, predictions))
```

```python
email = input("Enter Email Text: ")

email_vector = vectorizer.transform([email])

prediction = model.predict(email_vector)

print("\nPrediction:", prediction[0])

if prediction[0] == "Phishing":
    print("Warning: This email appears to be a phishing attempt.")
else:
    print("This email appears to be safe.")
```
